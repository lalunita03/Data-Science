```python
# import packages
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd 
import requests
import time


path = '/Users/lea/desktop/chromedriver-mac-arm64/chromedriver'

option = webdriver.ChromeOptions()
service = Service(executable_path=path)
driver = webdriver.Chrome(options = option)

# Step 1
driver.get("https://en.wikipedia.org/wiki/Main_Page")

# Step 2
search_bar = driver.find_element('name', 'search')
search_bar.send_keys('Caltech\n')

# Step 3
link = driver.find_element(By.LINK_TEXT, 'SCIAC')
link.click()

# Step 4
timeline = driver.find_element(By.ID, 'mw-content-text')
list = timeline.find_elements(By.TAG_NAME, 'ul')[2]
events = list.find_elements(By.TAG_NAME, 'li')
```

```python
# Step 4 cont.
years = []
description = []
for e in events:
    splits = e.text.split(" - ")
    years.append(splits[0])
    description.append(splits[1])

data = pd.DataFrame({"years": years, "description": description})

display(data)
```

```python
# Step 5
driver.back()

# Step 6
figure = driver.find_elements(By.TAG_NAME, 'figure')[4]
img = figure.find_element(By.TAG_NAME, 'img')
img_link = img.get_attribute('src')

response = requests.get(img_link).content
with open('/Users/lea/desktop/hw5-img.jpg', 'wb') as handler:
    handler.write(response)
```

```python
# Step 7
driver.get("https://tineye.com/")

# Step 8 (check the "I'm not a robot." box manually)
upload_box = driver.find_element(By.ID, 'upload_box')
upload_box.send_keys('/Users/lea/desktop/hw5-img.jpg')
```

```python
# Step 9
urls = []

pagination = driver.find_element(By.CLASS_NAME, 'pagination')
pages = pagination.find_elements(By.TAG_NAME, 'a')
n_pages = pages[-2].text

for i in range(int(n_pages)):
    time.sleep(1)
    elements = driver.find_elements(By.CLASS_NAME, 'match')
    for e in elements:
        h4 = e.find_element(By.TAG_NAME, 'h4')
        a = h4.find_element(By.TAG_NAME, 'a')
        urls.append(a.get_attribute('href'))
    
    # get next page
    next = driver.find_element(By.CLASS_NAME, 'next')
    if next is None:
        print("Somethign went wrong")
        break
    next.click()
    
display(urls)
```
