```python
## Import necessary packages
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
## Part 1

# Obtain and input into a Pandas dataframe the title (column 1), price (column 2), 
# and the number of stars in the rating (column 3) of the first 20 books that show up when you open up the website.

url = 'https://books.toscrape.com/'
response =requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
articles = soup.find_all('article', class_='product_pod')

data = {"title": [], "price": [], "rating": []}

for i in range(20):
    data['title'].append(articles[i].h3.a.get_text())
    data['price'].append(articles[i].find('p', class_='price_color').get_text())
    data['rating'].append(articles[i].find('p', class_='star-rating')['class'][1])

books = pd.DataFrame(data)

display(books)
```

```python
# Part 3
#Finally, find and export into one list all the UPCs (under Product Information) 
# of books that have the word “Boys” in the title.
upcs = []

url = 'https://books.toscrape.com/'

next_btn = ''
next = url

while (next_btn != None): # check if next exists
    if (next.endswith('page-2.html')):
        url = next[:len(next) - len('page-2.html')]
        
    if (next_btn != ''):
        next = url + next_btn.a['href']

    response =requests.get(next)
    soup = BeautifulSoup(response.content, 'html.parser')
    articles = soup.find_all('article', class_='product_pod')

    for i in range(len(articles)):
        # use link because title doesn't display entire title for long titles
        link = articles[i].h3.a['href']

        if ("boys" in link):
            url_i = url + link
            response_i = requests.get(url_i)
            soup_i = BeautifulSoup(response_i.content, 'html.parser')
            title = soup_i.find('div', class_='product_main').h1.get_text()
            print(title)
            table = soup_i.find('table', class_='table table-striped')
            upcs.append(table.tr.td.get_text())
            
    next_btn = soup.find('li', class_='next')

print(upcs)
```
